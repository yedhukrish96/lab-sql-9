```python
#Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
query = '''SELECT rental_id, rental_date, DAYNAME(rental_date) as day_of_week, 
CASE 
    WHEN DAYOFWEEK(rental_date) IN (1,7) THEN 'weekend'
    ELSE 'workday'
END AS day_type
FROM rental;'''
print(query)
data1 = pd.read_sql_query(query, engine)
data1.head(20)
```

```python
#Get all films with ARMAGEDDON in the title.
query = r'''SELECT title
FROM film
WHERE title like "%%ARMAGEDDON" OR title LIKE "ARMAGEDDON%%";'''   #Use two % sign to avoid error
print(query)
data2 = pd.read_sql_query(query, engine)
data2.head(20)

```

```python
#Get 10 the longest films.
query = '''SELECT MAX(length)
FROM film
LIMIT 10;'''
print(query)
data2 = pd.read_sql_query(query, engine)
data2.head(20)
```

```python
#How many films include Behind the Scenes content?
query = r'''SELECT special_features
FROM film
WHERE special_features LIKE "%%Behind the Scenes" OR special_features LIKE "Behind the Scenes%%";'''
print(query)
data2 = pd.read_sql_query(query, engine)
data2.head(20)
```

```python
#Which kind of movies (rating) have a mean duration of more than two hours?
```

```python
query = '''SELECT rating, ROUND(AVG(length),2) AS avg_length
FROM film
GROUP BY rating
HAVING AVG(length)>120;'''
print(query)
data2 = pd.read_sql_query(query, engine)
data2.head(20)
```

```python
#Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
query = '''SELECT title, length, RANK() OVER (ORDER BY length DESC) AS rank_no
FROM film
WHERE length IS NOT NULL AND length > 0
ORDER BY length DESC;'''
print(query)
data2 = pd.read_sql_query(query, engine)
data2.head(20)
```
